```python
import geemap,geemap as import_map
import ee
ee.Authenticate()
ee.Initialize(project='debayan2002')
vis_map=import_map.Map()
```

```python
studyarea=ee.FeatureCollection('projects/debayan2002/assets/study_area/Visakhapatnam_Boundary') #defining my study area
```

```python
landsat7_data = ee.ImageCollection("LANDSAT/LE07/C02/T1").filterDate('2000-01-01', '2000-12-31').filterMetadata('CLOUD_COVER', 'less_than', 10)
landsat9_data = ee.ImageCollection("LANDSAT/LC09/C02/T1").filterDate('2024-01-01', '2024-12-31').filterMetadata('CLOUD_COVER', 'less_than', 5) #filtering criteria for datasets
```

```python
landsat7_vis_parameter_fcc={'min':15.5,'max':142.5}
landsat7_vis_parameter_tcc={'min':33.5,'max':158} #setting the visual parametets for landsat 7 TCC and FCC data (2000)
```

```python
landsat7_fcc_2000=landsat7_data.select('B4','B3','B2').mean().clip(studyarea)
landsat7_tcc_2000=landsat7_data.select('B3','B2','B1').mean().clip(studyarea)
vis_map.addLayer(landsat7_fcc_2000,landsat7_vis_parameter_fcc,'FCC_2000')
vis_map.addLayer(landsat7_tcc_2000,landsat7_vis_parameter_tcc,'TCC_2000') #adding TCC and FCC datasets of Landsat 7 for the year of 2000
```

```python
landsat9_vis_parameter_fcc={'min':7482,'max':18443}
landsat9_vis_parameter_tcc={'min':7482,'max':14098}  #setting the visual parametets for landsat 9 TCC and FCC data (2024)
```

```python
landsat9_fcc_2024=landsat9_data.select('B5','B4','B3').mean().clip(studyarea)
landsat9_tcc_2024=landsat9_data.select('B4','B3','B2').mean().clip(studyarea)
vis_map.addLayer(landsat9_fcc_2024,landsat9_vis_parameter_fcc,'FCC_2024')
vis_map.addLayer(landsat9_tcc_2024,landsat9_vis_parameter_tcc,'TCC_2024') #adding TCC and FCC datasets of Landsat 9 for the year of 2024
```

```python
vis_map.centerObject(studyarea,9)
vis_map
```
