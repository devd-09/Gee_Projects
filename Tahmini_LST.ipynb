```python
import geemap,geemap as import_map
import ee
import os
ee.Authenticate()
ee.Initialize(project='debayan2002')
vis_map=import_map.Map()
```

```python
study_area=ee.FeatureCollection('projects/debayan2002/assets/mulshi_fieldwork/Tahmini_Boundary')
vis_map.addLayer(study_area,{},'Study Area')
vis_map.centerObject(study_area,10)
```

```python
# Function to apply scaling factors
def apply_scale_factors(image):
    # Scale and offset values for optical bands
    optical_bands = image.select('SR_B.*').multiply(0.0000275).add(-0.2)

    # Scale and offset values for thermal bands
    thermal_bands = image.select('ST_B.*').multiply(0.00341802).add(149.0)

    # Add scaled bands back to the original image
    return image.addBands(optical_bands, None, True).addBands(thermal_bands, None, True)
```

```python
# Function to Mask Clouds and Cloud Shadows in Landsat 8 Imagery
def cloud_mask(image):
    # Define cloud shadow and cloud bitmasks (Bits 3 and 5)
    cloud_shadow_bitmask = (1 << 3)
    cloud_bitmask = (1 << 5)

    # Select the Quality Assessment (QA) band for pixel quality information
    qa = image.select('QA_PIXEL')

    # Create a binary mask to identify clear conditions (both cloud and cloud shadow bits set to 0)
    mask = qa.bitwiseAnd(cloud_shadow_bitmask).eq(0).And(qa.bitwiseAnd(cloud_bitmask).eq(0))

    # Update the original image, masking out cloud and cloud shadow-affected pixels
    return image.updateMask(mask)
```

```python
landsat5_1990=ee.ImageCollection("LANDSAT/LT05/C02/T1_L2").filterBounds(study_area).filterDate('1990-01-01','1990-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #1990
landsat5_1995=ee.ImageCollection("LANDSAT/LT05/C02/T1_L2").filterBounds(study_area).filterDate('1995-01-01','1995-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #1995
landsat7_2000=ee.ImageCollection("LANDSAT/LE07/C02/T1_L2").filterBounds(study_area).filterDate('2000-01-01','2000-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #2000
landsat7_2005=ee.ImageCollection("LANDSAT/LE07/C02/T1_L2").filterBounds(study_area).filterDate('2005-01-01','2005-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #2005
landsat7_2010=ee.ImageCollection("LANDSAT/LE07/C02/T1_L2").filterBounds(study_area).filterDate('2010-01-01','2010-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #2010
landsat8_2015=ee.ImageCollection("LANDSAT/LE07/C02/T1_L2").filterBounds(study_area).filterDate('2015-01-01','2015-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #2015
landsat8_2020=ee.ImageCollection("LANDSAT/LC08/C02/T1_L2").filterBounds(study_area).filterDate('2020-01-01','2020-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #2020
landsat9_2024=ee.ImageCollection("LANDSAT/LC09/C02/T1_L2").filterBounds(study_area).filterDate('2024-01-01','2024-12-31').map(apply_scale_factors).map(cloud_mask).mean().clip(study_area) #2024
```

```python
# Calculate the minimum NDVI value within the AOI
ndvi_min_1990 = ndvi_1990.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
ndvi_min_1995 = ndvi_1995.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
ndvi_min_2000 = ndvi_2000.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
ndvi_min_2005 = ndvi_2005.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
ndvi_min_2010 = ndvi_2010.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
ndvi_min_2015 = ndvi_2015.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
ndvi_min_2020 = ndvi_2020.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
ndvi_min_2024 = ndvi_2024.reduceRegion(reducer=ee.Reducer.min(),geometry=study_area,scale=30,maxPixels=1e9).values().get(0)
```

```python
ndvi_max_1990_img = ee.Image.constant(ndvi_max_1990)
ndvi_min_1990_img = ee.Image.constant(ndvi_min_1990)
ndvi_max_1995_img = ee.Image.constant(ndvi_max_1995)
ndvi_min_1995_img = ee.Image.constant(ndvi_min_1995)
ndvi_max_2000_img = ee.Image.constant(ndvi_max_2000)
ndvi_min_2000_img = ee.Image.constant(ndvi_min_2000)
ndvi_max_2005_img = ee.Image.constant(ndvi_max_2005)
ndvi_min_2005_img = ee.Image.constant(ndvi_min_2005)
ndvi_max_2010_img = ee.Image.constant(ndvi_max_2010)
ndvi_min_2010_img = ee.Image.constant(ndvi_min_2010)
ndvi_max_2015_img = ee.Image.constant(ndvi_max_2015)
ndvi_min_2015_img = ee.Image.constant(ndvi_min_2015)
ndvi_max_2020_img = ee.Image.constant(ndvi_max_2020)
ndvi_min_2020_img = ee.Image.constant(ndvi_min_2020)
ndvi_max_2024_img = ee.Image.constant(ndvi_max_2024)
ndvi_min_2024_img = ee.Image.constant(ndvi_min_2024)
```

```python
# Fraction of Vegetation (FV) Calculation
fv_1990= ((ndvi_1990.subtract(ndvi_min_1990_img)).divide(ndvi_max_1990_img.subtract(ndvi_min_1990_img))).pow(2).rename('FV_1990')
fv_1995= ((ndvi_1995.subtract(ndvi_min_1995_img)).divide(ndvi_max_1995_img.subtract(ndvi_min_1995_img))).pow(2).rename('FV_1995')
fv_2000= ((ndvi_2000.subtract(ndvi_min_2000_img)).divide(ndvi_max_2000_img.subtract(ndvi_min_2000_img))).pow(2).rename('FV_2000')
fv_2005= ((ndvi_2005.subtract(ndvi_min_2005_img)).divide(ndvi_max_2005_img.subtract(ndvi_min_2005_img))).pow(2).rename('FV_2005')
fv_2010= ((ndvi_2010.subtract(ndvi_min_2010_img)).divide(ndvi_max_2010_img.subtract(ndvi_min_2010_img))).pow(2).rename('FV_2010')
fv_2015= ((ndvi_2015.subtract(ndvi_min_2015_img)).divide(ndvi_max_2015_img.subtract(ndvi_min_2015_img))).pow(2).rename('FV_2015')
fv_2020= ((ndvi_2020.subtract(ndvi_min_2020_img)).divide(ndvi_max_2020_img.subtract(ndvi_min_2020_img))).pow(2).rename('FV_2020')
fv_2024= ((ndvi_2024.subtract(ndvi_min_2024_img)).divide(ndvi_max_2024_img.subtract(ndvi_min_2024_img))).pow(2).rename('FV_2024')
```

```python
# Emissivity (EM) Calculation
em_1990 = fv_1990.multiply(0.004).add(0.986).rename('EM_1990')
em_1995 = fv_1995.multiply(0.004).add(0.986).rename('EM_1995')
em_2000 = fv_2000.multiply(0.004).add(0.986).rename('EM_2000')
em_2005 = fv_2005.multiply(0.004).add(0.986).rename('EM_2005')
em_2010 = fv_2010.multiply(0.004).add(0.986).rename('EM_2010')
em_2015 = fv_2015.multiply(0.004).add(0.986).rename('EM_2015')
em_2020 = fv_2020.multiply(0.004).add(0.986).rename('EM_2020')
em_2024 = fv_2024.multiply(0.004).add(0.986).rename('EM_2024')
```
