```python
import ee
import geemap
from IPython.display import display, HTML
from PIL import Image
!pip install ffmpeg-python
import os
import base64
```

```python
ee.Authenticate()
ee.Initialize(project="debayan2002")
vis_map = geemap.Map()
vis_map.add_basemap("SATELLITE")
vis_map
```

```python
roi = vis_map.user_roi
if roi is None:
    print("Please draw a polygon on the map to set the ROI.")
else:
    def mask_clouds(image):
        qa = image.select('QA_PIXEL')
        cloud_mask = qa.bitwiseAnd(1 << 4).eq(0)
        return image.updateMask(cloud_mask)
```

```python
out_gif = 'landsat8_timelapse.gif'
out_mp4 = out_gif.replace('.gif', '.mp4')
if os.path.exists(out_mp4):
   os.remove(out_mp4)
```

```python
geemap.landsat_timelapse(
        roi=roi,
        out_gif=out_gif,
        start_year=2000,
        end_year=2024,
        start_date='01-01',
        end_date='12-31',
        bands=['NIR', 'SWIR1', 'Red'],
        dimensions=700,
        title='Landsat 8 Temporal Changes (2000-2024)',
        progress_bar_color='blue',
        mp4=True,
        frames_per_second=10,
    )
```

```python
def display_video_mp4(video_path):
        with open(video_path, 'rb') as f:
            video_base64 = base64.b64encode(f.read()).decode()
        video_html = f'''
        <video width="600" controls loop>
            <source src="data:video/mp4;base64,{video_base64}" type="video/mp4">
        </video>
        '''
        return HTML(video_html)
        if os.path.exists(out_gif):
            display(Image.open(out_gif))
        else:
            print(f"Error: The GIF file '{out_gif}' was not found.")
if os.path.exists(out_mp4):
        display(display_video_mp4(out_mp4))
else:
        print(f"Error: The MP4 file '{out_mp4}' was not found.")



```
