```python
import geemap,geemap as import_map
import ee
import ipyleaflet
ee.Authenticate()
ee.Initialize(project='debayan2002')
vis_map=import_map.Map()
vis_map2=import_map.Map()
vis_map3=import_map.Map()
```

```python
import ipywidgets as widgets
from ipywidgets import HTML
```

```python
study_area=ee.FeatureCollection('projects/debayan2002/assets/study_area/Visakhapatnam_Boundary')
vis_map.addLayer(study_area,{},'Study Area')
vis_map2.addLayer(study_area,{},'Study Area')
vis_map3.addLayer(study_area,{},'Study Area')
vis_map.centerObject(study_area,9)
vis_map2.centerObject(study_area,9)
vis_map3.centerObject(study_area,9)
```

```python
landsat5_1990=ee.ImageCollection("LANDSAT/LT05/C02/T1").filterDate('1990-01-01','1990-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #1990
landsat7_2000=ee.ImageCollection("LANDSAT/LE07/C02/T1").filterDate('2000-01-01','2000-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2000
landsat7_2010=ee.ImageCollection("LANDSAT/LE07/C02/T1").filterDate('2010-01-01','2010-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2010
landsat8_2020=ee.ImageCollection("LANDSAT/LC08/C02/T1").filterDate('2020-01-01','2020-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2020
landsat9_2024=ee.ImageCollection("LANDSAT/LC09/C02/T1").filterDate('2024-01-01','2024-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2024
```

```python
ndvi_legend_dict={
    "Barren Land / Water Bodies": "#C58047",
    "Built Ups": "#F3D0BD",
    "Sparse Vegetation": "#FDF9AF",
    "Moderate Vegetation": "#C7CF45",
    "Dense Vegetation": "#122200"}
```

```python
ndvi_palette=(['#8B4513', '#9A5120', '#A9602D', '#B8703A', '#C58047', '#D49055', '#E3A063', '#E9B081', '#EEC09F', '#F3D0BD', '#F8E0DB', '#FCF0F9', '#FCF3E1', '#FDF6C8', '#FDF9AF', '#FEFC96', '#F7F580', '#E7E86C', '#D7DC58', '#C7CF45', '#B7C332', '#A7B71F', '#97AA0D', '#889E0A', '#789209', '#698607', '#5A7A06', '#4B6E05', '#3C6203', '#2D5602', '#1F4A01', '#154000', '#133800', '#123000', '#112800', '#102000', '#101800', '#101000', '#101000', '#122200'])
```

```python
#NDVI from 1990 to 2024
ndvi_1990=landsat5_1990.normalizedDifference(['B4','B3'])
ndvi_2000=landsat7_2000.normalizedDifference(['B4','B3'])
ndvi_2010=landsat7_2010.normalizedDifference(['B4','B3'])
ndvi_2020=landsat8_2020.normalizedDifference(['B5','B4'])
ndvi_2024=landsat9_2024.normalizedDifference(['B5','B4'])
```

```python
ndwi_palette=(['#CEFFFD', '#76E3DE', '#76CCE3', '#40B4D3', '#4078D3', '#2126BB'])
```

```python
ndwi_legend_dict={
    "Extremely Dry": "#CEFFFD",
    "Dry": "#76CCE3",
    "Moderate": "#40B4D3",
    "Wet": "#4078D3",
    "Extremely Wet": "#2126BB"}
```

```python
#NDWI from 1990 to 2024
ndwi_1990=landsat5_1990.normalizedDifference(['B2','B4'])
ndwi_2000=landsat7_2000.normalizedDifference(['B2','B4'])
ndwi_2010=landsat7_2010.normalizedDifference(['B2','B4'])
ndwi_2020=landsat8_2020.normalizedDifference(['B3','B5'])
ndwi_2024=landsat9_2024.normalizedDifference(['B3','B5'])
```

```python
savi_legend_dict={
    "Barren Land / Water Bodies": "#5A2C0F",
    "Built Ups": "#D1A472",
    "Sparse Vegetation": "#FBE9C4",
    "Moderate Vegetation": "#D2F599",
    "Dense Vegetation": "#74E65B"}
```

```python
savi_palette=(['#5A2C0F', '#6A3A1C', '#7B4C28', '#8C5E34', '#9D7040', '#AE8250', '#BF9360', '#D1A472', '#E2B684', '#F3C896', '#F7DAA8', '#F9E1B6', '#FBE9C4', '#FFEECC', '#FFF4D2', '#F7F8B4', '#E5F6A6', '#D2F599', '#C0F38B', '#AEF17E', '#9DEF71', '#88EA66', '#74E65B', '#60E250', '#4EDD47', '#3DD93E', '#30CC3A', '#28B532', '#209E2B', '#188825', '#117320', '#0A5E1A', '#064A14', '#00360F'])
```

```python
savi_1990= landsat5_1990.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat5_1990.select('B4'),'RED': landsat5_1990.select('B3')})
savi_2000= landsat7_2000.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat7_2000.select('B4'),'RED': landsat7_2000.select('B3')})
savi_2010= landsat7_2010.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat7_2010.select('B4'),'RED': landsat7_2000.select('B3')})
savi_2020= landsat8_2020.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat8_2020.select('B5'),'RED': landsat8_2020.select('B4')})
savi_2024= landsat9_2024.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat9_2024.select('B5'),'RED': landsat9_2024.select('B4')})
```

```python
#ndvi add layer
vis_map.addLayer(ndvi_1990,{'min':-0.2,'max':0.5,'palette':ndvi_palette},'NDVI 1990')
vis_map.addLayer(ndvi_2000,{'min':-0.4,'max':0.5,'palette':ndvi_palette},'NDVI 2000')
vis_map.addLayer(ndvi_2010,{'min':-0.4,'max':0.3,'palette':ndvi_palette},'NDVI 2010')
vis_map.addLayer(ndvi_2020,{'min':-0.1,'max':0.4,'palette':ndvi_palette},'NDVI 2020')
vis_map.addLayer(ndvi_2024,{'min':0,'max':0.4,'palette':ndvi_palette},'NDVI 2024')
vis_map.addLayerControl()
vis_map.add_legend(title="NDVI Legend",legend_dict=ndvi_legend_dict,position="bottomright")
vis_map
```

```python
#ndwi add layer
vis_map2.addLayer(ndwi_1990,{'min':-0.4,'max':0.4,'palette':ndwi_palette},'NDWI 1990')
vis_map2.addLayer(ndwi_2000,{'min':-0.4,'max':0.4,'palette':ndwi_palette},'NDWI 2000')
vis_map2.addLayer(ndwi_2010,{'min':-0.2,'max':0.5,'palette':ndwi_palette},'NDWI 2010')
vis_map2.addLayer(ndwi_2020,{'min':-0.4,'max':0.1,'palette':ndwi_palette},'NDWI 2020')
vis_map2.addLayer(ndwi_2024,{'min':-0.3,'max':0.1,'palette':ndwi_palette},'NDWI 2024')
vis_map2.addLayerControl()
vis_map2.add_legend(title="NDWI Legend",legend_dict=ndwi_legend_dict,position="bottomright")
vis_map2
```

```python
#savi add layer
vis_map3.addLayer(savi_1990,{'min':-0.3,'max':0.8,'palette':savi_palette},'SAVI 1990')
vis_map3.addLayer(savi_2000,{'min':-0.6,'max':0.9,'palette':savi_palette},'SAVI 2000')
vis_map3.addLayer(savi_2010,{'min':-0.7,'max':0.7,'palette':savi_palette},'SAVI 2010')
vis_map3.addLayer(savi_2020,{'min':-0.1,'max':0.7,'palette':savi_palette},'SAVI 2020')
vis_map3.addLayer(savi_2024,{'min':-0.1,'max':0.6,'palette':savi_palette},'SAVI 2024')
vis_map3.addLayerControl()
vis_map3.add_legend(title="SAVI Legend",legend_dict=savi_legend_dict,position="bottomright")
vis_map3
```
