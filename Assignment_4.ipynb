```python
import geemap
import ee
import os
ee.Authenticate()
ee.Initialize(project='debayan2002')
```

```python
vis_map=geemap.Map()
```

```python
india_districts=ee.FeatureCollection('projects/debayan2002/assets/India_district')
vis_map.addLayer(india_districts,{},'India_district')#adding shapefile
```

```python
dsr_data=ee.ImageCollection("MODIS/061/MCD18A1") #adding the main dataset
```

```python
dsr_2010=dsr_data.filterDate('2010-01-01','2010-12-31').select('DSR').mean().clip(india_districts)
dsr_2015=dsr_data.filterDate('2015-01-01','2015-12-31').select('DSR').mean().clip(india_districts)
dsr_2020=dsr_data.filterDate('2020-01-01','2020-12-31').select('DSR').mean().clip(india_districts)
```

```python
dsr_palette=(['#006837', '#087145', '#108B54', '#19A562', '#22BF71', '#2BD980', '#34F38E', '#52F597',
'#70F7A1', '#8EF9AB', '#ACFDB6', '#CAFFC1', '#E8FFCC', '#FCF7BF', '#FCE88C', '#FCD959',
'#FCC92A', '#FDBA00', '#FCA300', '#FB8C00', '#F97500', '#F85E00', '#F64700'])
```

```python
#setting the min max unit of DSR
unit_2010={'min':342.6,'max':790.3,'palette':dsr_palette}
unit_2015={'min':329.6,'max':800.7,'palette':dsr_palette}
unit_2020={'min':315.8,'max':786.9,'palette':dsr_palette}
```

```python
vis_map.addLayer(dsr_2015,unit_2015,'DSR of 2015')
vis_map.addLayer(dsr_2020,unit_2020,'DSR of 2020')
vis_map.addLayer(dsr_2010,unit_2010,'DSR of 2010')
vis_map
```

```python
output_path=os.path.join(os.path.expanduser("~"), "Downloads")
out_dsr_stats = os.path.join(output_path, "dsr_2010.csv")
if not os.path.exists(output_path):
    os.makedirs(output_path)
geemap.zonal_stats(dsr_2010,india_districts, out_dsr_stats, stat_type="MEAN", scale=250) #download csv for dsr 2010
```

```python
output_path=os.path.join(os.path.expanduser("~"), "Downloads")
out_dsr_stats = os.path.join(output_path, "dsr_2015.csv")
if not os.path.exists(output_path):
    os.makedirs(output_path)
geemap.zonal_stats(dsr_2015,india_districts, out_dsr_stats, stat_type="MEAN", scale=250) #download csv for dsr 2015
```
