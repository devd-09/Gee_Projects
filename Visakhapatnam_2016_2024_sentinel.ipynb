```python
import geemap,geemap as import_map
import ee
ee.Authenticate()
ee.Initialize(project='debayan2002')
vis_map=import_map.Map()
```

```python
studyarea=ee.FeatureCollection('projects/debayan2002/assets/study_area/Visakhapatnam_Boundary') #defining my study area
```

```python
sentinel_data_2016=ee.ImageCollection("COPERNICUS/S2_HARMONIZED").filterDate('2016-01-01', '2016-12-31').filterMetadata('CLOUD_COVERAGE_ASSESSMENT', 'less_than', 5)
sentinel_data_2024=ee.ImageCollection("COPERNICUS/S2_HARMONIZED").filterDate('2024-01-01', '2024-12-31').filterMetadata('CLOUD_COVERAGE_ASSESSMENT', 'less_than', 5)
```

```python
sentinel_2016_tcc=sentinel_data_2016.select('B4','B3','B2').mean().clip(studyarea)
sentinel_2016_fcc=sentinel_data_2016.select('B8','B4','B3').mean().clip(studyarea)
```

```python
sentinel_2024_tcc=sentinel_data_2024.select('B4','B3','B2').mean().clip(studyarea)
sentinel_2024_fcc=sentinel_data_2024.select('B8','B4','B3').mean().clip(studyarea)
```

```python
sentinel2016_tcc_vis_para={'min':610.3,'max':1742.5}
sentinel2016_fcc_vis_para={'min':456.5,'max':3636}
sentinel_2024_tcc_vis_para={'min':610.3,'max':1717.5}
sentinel_2024_fcc_vis_para={'min':487.4,'max':2896.8}
```

```python
vis_map.addLayer(sentinel_2016_tcc,sentinel2016_tcc_vis_para,'TCC_2016')
vis_map.addLayer(sentinel_2016_fcc,sentinel2016_fcc_vis_para,'FCC_2016')
vis_map.addLayer(sentinel_2024_tcc,sentinel_2024_tcc_vis_para,'TCC_2024')
vis_map.addLayer(sentinel_2024_fcc,sentinel_2024_fcc_vis_para,'FCC_2024')
```

```python
vis_map.center_object(studyarea,9)
vis_map
```
