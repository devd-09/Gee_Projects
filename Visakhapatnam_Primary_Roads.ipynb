```python
import geemap,geemap as import_map
import ee
ee.Authenticate()
ee.Initialize(project='debayan2002')
vis_map=import_map.Map()
```

```python
studyarea=ee.FeatureCollection('projects/debayan2002/assets/study_area/Visakhapatnam_Boundary')
roadway=ee.FeatureCollection('projects/debayan2002/assets/study_area/Roadways')
primary_roads=roadway.filter(ee.Filter.eq('fclass','primary'))
```

```python
studyarea_style = {'color': 'black', 'fillOpacity': 0.1}
roadway_style = {'color': 'red', 'width': 2, 'lineType': 'solid'}
```

```python
vis_map.addLayer(studyarea,studyarea_style,'Boundary')
vis_map.addLayer(primary_roads,roadway_style,'Primary Roads')
vis_map.centerObject(studyarea,9)
vis_map
```
